#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

```python
# para hacer 4 scatter plots de los precios y la superficie de cada poblacion de mi df2 y usando subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].scatter(df2[df2['level5'] == 'Fuenlabrada']['surface'], df2[df2['level5'] == 'Fuenlabrada']['price'])
axs[0, 0].set_title('Fuenlabrada')
axs[0, 1].scatter(df2[df2['level5'] == 'Leganés']['surface'], df2[df2['level5'] == 'Leganés']['price'])
axs[0, 1].set_title('Leganés')
axs[1, 0].scatter(df2[df2['level5'] == 'Getafe']['surface'], df2[df2['level5'] == 'Getafe']['price'])
axs[1, 0].set_title('Getafe')
axs[1, 1].scatter(df2[df2['level5'] == 'Alcorcón']['surface'], df2[df2['level5'] == 'Alcorcón']['price'])
axs[1, 1].set_title('Alcorcón')
for ax in axs.flat:
    ax.set(xlabel='Superficie', ylabel='Precio')
plt.tight_layout()
plt.show()
#Se evidencia que la relación entre superficie y precio es lineal en todas las poblaciones
```

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps

# Mapa centrado en Madrid
map = Map(center = (40.35, -3.8), zoom = 11.5, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
from ipyleaflet import Map, Marker

coordenadas = df2[['latitude', 'longitude']].to_dict(orient='records')

m = Map(center=(60,-2.2), zoom=4, min_zoom=1, max_zoom=20)

for coord in coordenadas:
    marker = Marker(location=(coord['latitude'], coord['longitude']))
    m.add_layer(marker)
    
m
```
